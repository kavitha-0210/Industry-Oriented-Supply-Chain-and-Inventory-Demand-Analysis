```python
import pandas as pd
import numpy as np

data = {
    'Month':['Jan','Feb','Mar','Apr','May','Jun',
             'Jul','Aug','Sep','Oct','Nov','Dec'],
    'Demand':[120,135,140,160,180,200,
              220,210,230,250,270,300],
    'Inventory':[150,160,170,180,190,210,
                 250,240,260,280,300,320],
    'Sales':[110,130,135,155,175,190,
             210,200,220,240,260,290]
}

df = pd.DataFrame(data)

print(df)
```

```python
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```python
df['Inventory_Turnover'] = (
    df['Sales'] /
    df['Inventory']
)

print(df[['Month','Inventory_Turnover']])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

plt.plot(df['Month'],
         df['Demand'],
         marker='o')

plt.title("Demand Trend Analysis")
plt.xlabel("Month")
plt.ylabel("Demand")

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(df['Month'],
         df['Inventory'],
         marker='o')

plt.title("Inventory Levels")
plt.xlabel("Month")
plt.ylabel("Inventory")

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(df['Month'],
         df['Demand'],
         label='Demand')

plt.plot(df['Month'],
         df['Sales'],
         label='Sales')

plt.legend()

plt.title("Sales vs Demand")
plt.show()
```

```python
df['Stock_Out'] = np.where(
    df['Demand'] > df['Inventory'],
    1,
    0
)

print(df[['Month','Stock_Out']])
```

```python
df['Overstock'] = np.where(
    df['Inventory'] > (df['Demand'] * 1.2),
    1,
    0
)

print(df[['Month','Overstock']])
```

```python
from sklearn.linear_model import LinearRegression

df['Month_No'] = range(1,13)

X = df[['Month_No']]
y = df['Demand']

model = LinearRegression()

model.fit(X,y)

future = pd.DataFrame({
    'Month_No':[13,14,15]
})

forecast = model.predict(future)

print("Future Demand Forecast")
print(forecast)
```

```python
future_months = ['Jan_Next',
                 'Feb_Next',
                 'Mar_Next']

plt.figure(figsize=(10,5))

plt.plot(df['Month'],
         df['Demand'],
         marker='o',
         label='Historical Demand')

plt.plot(future_months,
         forecast,
         marker='o',
         linestyle='--',
         label='Forecast')

plt.legend()

plt.title("Demand Forecast")
plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.bar(df['Month'],
        df['Inventory_Turnover'])

plt.title("Inventory Turnover Analysis")
plt.xlabel("Month")
plt.ylabel("Turnover")

plt.show()
```
